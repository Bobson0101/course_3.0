```python
import pandas as pd
import numpy as np
```

```python
data_Xy = {'興趣':['看電視','讀書','音樂','看電視'],'成功與否':['是','否','否','是']}
df = pd.DataFrame(data = data_Xy, index=['小明','小林','小英','小陳'])
df = df[['興趣','成功與否']]
df
```

```python
X = pd.get_dummies(df.iloc[:, 0])  # 針對第一個欄位做get_dummies
X
```

```python
X = pd.get_dummies(df.iloc[:, 1])  # 針對第一個欄位做get_dummies
X
```
